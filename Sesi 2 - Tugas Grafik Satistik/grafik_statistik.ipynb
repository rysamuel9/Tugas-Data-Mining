```python
# MULTIPEL HISTOGRAM

n_bins = 20
# x = np.random.randn(1000, 3)
nilai_tugas = dataset[['Nilai Tugas', 'Nilai UTS', 'Nilai UAS']]
# nilai_uts = dataset['Nilai UTS']
# nilai_uas = dataset['Nilai UAS']

fig, ax0 = plt.subplots(nrows=1, ncols=1)

colors = ['red', 'blue', 'green']
labels = ['Nilai Tugas', 'Nilau UTS', 'Nilai UAS']
ax0.hist(nilai_tugas, n_bins, density=True, histtype='bar', color=colors, label=labels)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')
```
